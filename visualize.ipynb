```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("books_data_cleaned.csv")   

# Check column names to ensure they match your dataset
print(df.columns)

sns.set(style="whitegrid", palette="muted")

# Distribution of Book Prices
plt.figure(figsize=(6,4))
sns.histplot(df["Price"], kde=True, color="teal", edgecolor="black")
plt.title("Distribution of Book Prices")
plt.xlabel("Price")
plt.ylabel("Count")
plt.savefig("price_distribution.png", dpi=300, bbox_inches='tight')
plt.show()

# Book Ratings Distribution
plt.figure(figsize=(6,4))
sns.countplot(x="Rating", data=df, palette="muted")
plt.title("Book Ratings Distribution")
plt.savefig("rating_distribution.png", dpi=300, bbox_inches='tight')
plt.show()

# Price vs Rating
plt.figure(figsize=(6,4))
sns.boxplot(x="Rating", y="Price", data=df, palette="coolwarm")
plt.title("Price vs Rating")
plt.savefig("price_vs_rating.png", dpi=300, bbox_inches='tight')
plt.show()
```
